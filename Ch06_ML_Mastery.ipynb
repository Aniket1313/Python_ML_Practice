```
filename = 'pima-indians-diabetes.csv'
names = ['preg','plans','skin','test','mass','pedi','age','class']
df = pd.read_csv(filename,names=names)
```

```
df.head()
```

```
#6.1 Histogram

df.hist()
plt.show()
```

```
#6.1.2 Density plot :Gives distribution of each attribute
plt.figure(figsize=(30,15))
df.plot(kind='density',subplots=True,layout=(3,3,),sharex=False)
plt.show()
#We can see that distribution for each attribute is clearer than the histogram
```

```
#6.1.3 Box snd Whisker plots
#Box and Whisker plots 
#Whiskers give idea of spread of data and dots

df.plot(kind='box', subplots = True ,layout=(3,3),sharex = False ,sharey = False)
plt.figure(figsize=(10,15))
plt.show()
```

```
#6.2.1 Correlation Matrix Plot

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

```

```
correlations = df.corr()

#plot correlation matrix 

fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations,vmin =-1,vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```

```
#We can see symmetrical matrix.Bottom left is like top right.

#Scatter Plot Matrix :

pd.plotting.scatter_matrix(df, alpha=0.2)
plt.show()
```
