# Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
```

```python
tips.head()
```

## Scatter Plot

```python
import random

df = pd.DataFrame()

df['x'] = random.sample(range(1, 1000),100)
df['y'] = random.sample(range(1, 1000), 100)
df['z'] = np.random.randint(2, size=100)
```

```python
sns.scatterplot(x="x", y="y", data=df);
```

```python
sns.relplot(x="x", y="y", data=df, kind = 'scatter');
```

```python
sns.scatterplot(x="x", y="y", hue = "z", data=df);
```

```python
sns.relplot(x="x", y="y", hue = "z", data=df, kind = 'scatter');
```

#### Plotting Framework
1. Hypothesis or Question
2. Plot
3. Interpretation
4. More hypothesis or questions

#### Plotting Framework
1. Hypothesis or Question: Possible relationship between tip & bill
2. Plot: plotting bill against tip
3. Interpretation
4. More hypothesis or questions

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
sns.scatterplot(x="total_bill", y="tip", alpha=0.3,
                     data=tips)
```

#### Plotting Framework
1. Hypothesis or Question: Possible relationship between tip & bill
2. Plot: plotting bill against tip
3. Interpretation: Most tips seem to be proportional to bill
4. More hypothesis or questions: 
    1. Possible difference between tip percentage by country? by hotel type? 
    2. Possible difference of tip distribution by other categories? Gender or Time? 

```python
sns.scatterplot(x="total_bill", y="tip", hue="time",
                     data=tips)
```

```python
sns.scatterplot(x="total_bill", y="tip",
                     hue="day", style="time", data=tips)
```

```python
sns.scatterplot(x="total_bill", y="tip", size="size",
                     data=tips)
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="size",
                     data=tips)
```

```python
sns.relplot(x="total_bill", y="tip",
                 col="time", hue="day", style="day",
                 kind="scatter", data=tips)
```

```python
sns.scatterplot(x="total_bill", y="tip",
                 col="time", hue="day", style="day",
                 data=tips);
```

```python
sns.scatterplot(x=iris.sepal_length, y=iris.sepal_width,
                     hue=iris.species)
```
