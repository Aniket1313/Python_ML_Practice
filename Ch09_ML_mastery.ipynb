```
names = ['preg','plas','pres','skin','test','mass','pedi','age','class']
```

# 9.2 Split into Train and Test splits

```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

df = pd.read_csv('pima-indians-diabetes.csv',names = names)
array = df.values


X = array[:,0:8]
Y = array[:,8]
test_size = 0.33
seed =7
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=test_size,random_state = seed)
model = LogisticRegression()
model.fit(X_train,Y_train)
result = model.score(X_test,Y_test)
print("Accuracy: ", result*100.0)
```

# 9.3 K-Fold Cross Validation

It is an approach we can use to estimate the performance of a ML Algorithm with less var than a single train-test split.splits data into k-parts(k=5 or k=10)

```
from sklearn.model_selection import KFold,cross_val_score
```

```
#Evaluate using cross validation
num_folds = 10
seed =7
kfold = KFold(n_splits=num_folds ,random_state = seed)
model = LogisticRegression()
results = cross_val_score(model,X,Y,cv=kfold)
print("Accuracy : ", result.mean()*100.0, results.std()*100.0)
```

#9.6 What techniques to use when:


*   Generally k-fold cross validation is **gold standard** for evaluating the performance of a machine learning algorithm on unseen data with k set to 3,5,10
*   Use a train/test split is good for speed when using a slow algoritm produces performance estimates with lower bias when using large datasets.
*    Leave-one out CV and repeated random splits can be useful intermediates when trying to balance variance in estimated performance,model training speed and dataset size.

Use 10-fold cross validation,when in doubt.
